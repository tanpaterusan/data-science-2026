```python
import pandas as pd
import numpy as np
```

```python
# 1 Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)
```

```python
# 2 Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())
```

Kuis 1:
Berapa jumlah baris dan kolom dataset Titanic?

> 891 baris dan 12 kolom.




Kuis 2: Kolom mana yang memiliki nilai kosong (missing)?


> Age sebanyak 177 baris, Cabin sebanyak 687 baris dan Embarked sebanyak 2 baris.



```python

# 3 Filtering — Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') &
         (df['Pclass']==1)]
print(wk1['Survived'].mean(), wk1['Age'].mean())
```

Kuis 3: Rata-rata usia penumpang wanita kelas 1 yang selamat?


> 34 tahun.





```python
# 4 Groupby — Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat=('Survived','mean'),
).round(3)
print(hasil)
```

Kuis 4: Kelas penumpang mana yang tingkat keselamatan tertinggi?


> Penumpang kelas 3



```python
# Kuis 5: Tampilkan hasil groupby('Pclass')['Survived'].mean()

print(df.groupby('Pclass')['Survived'].mean())
```
